```python
import requests
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_excel("RPM Dataset.xlsx")
pd.options.display.max_rows=9999
```

```python
df.head(100)
```

```python
# Preprocess the data
# Convert the data types of other columns as needed
df['Temperature'] = df['Temperature'].astype(float)
df['Heart Rate'] = df['Heart Rate'].astype(float)
df['Pulse'] = df['Pulse'].astype(float)
df['Respiratory Rate'] = df['Respiratory Rate'].astype(float)
df['Oxygen Saturation'] = df['Oxygen Saturation'].astype(float)
```

```python
def predict_body_condition(df1):
# Feature Engineering
    # Create a new column for body condition
    conditions = []
    for index, row in df1.iterrows():
        if row['Temperature'] > 98:
            conditions.append('Severe')
        elif (row['Heart Rate'] > 100) or (row['Heart Rate'] < 60):
            conditions.append('Severe')
        elif (row['Pulse'] > 100) or (row['Pulse'] < 60):
            conditions.append('Severe')
        elif (row['Respiratory Rate'] > 16) or (row['Respiratory Rate'] < 12):
            conditions.append('Severe')
        elif (row['Oxygen Saturation'] < 0.95):
            conditions.append('Severe')
        else:
            conditions.append('Normal')
    df1['bodyCondition']=conditions
    
    #df1['body_condition'] = np.where(df1['Temperature'] > 98, 'Severe', 'Normal')
    #df1['body_condition'] = np.where((df1['Heart Rate'] > 100 )& (df1['Heart Rate'] < 60), 'Severe', 'Normal')
    #df1['body_condition'] = np.where((df1['Pulse'] > 100) & (df1['Pulse'] < 60), 'Severe', 'Normal')
    #df1['body_condition'] = np.where((df1['Respiratory Rate'] > 16) & (df1['Respiratory Rate'] < 12) ,'Severe' , 'Normal')
    #df1['body_condition'] = np.where((df1['Oxygen Saturation'] > 1.00) & (df1['Oxygen Saturation'] < 0.95) , 'Severe', 'Normal')

    # Select relevant features
    X = df1[['Temperature', 'Heart Rate','Pulse','Respiratory Rate','Oxygen Saturation']]
    y = df1['body_condition']
    # Train a Random Forest Classifier
    
    rfc = RandomForestClassifier(n_estimators=100, random_state=42)
    rfc.fit(X, y)
    # Predict the body condition for the latest data point
    latest_data = df1.iloc[-1][['Temperature', 'Heart Rate','Pulse','Respiratory Rate','Oxygen Saturation']]
    prediction = rfc.predict(latest_data.to_numpy().reshape(1, -1))[0]
    return prediction
```
